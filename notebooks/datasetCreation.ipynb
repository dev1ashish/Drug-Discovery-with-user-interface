```python
# Import necessary libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

## **Search for Target protein**

### **Target search for Acetylcholinesterase**

```python
# Target search for coronavirus
target = new_client.target
target_query = target.search('coronavirus')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
selected_target = targets.target_chembl_id[6]
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df= pd.DataFrame.from_dict(res)
df.head(3)
```

lower the number of standard value the better the potency of the drug

```python
df.standard_type.unique()
```

```python
df.to_csv('coronavirus_data.csv', index=False)
```

# handeling missing data

drop compounds with missing standard value

```python
df2 = df[df.standard_value.notna()]
df2
```

# data preprocessing

labelling compounds as either being active(less than 1000nM), intermidiate(bw active and inactive), inactive(greater than 10000 nM)

```python
bioactivity_class = []
for i in df2.standard_value:
    if float(i) >= 10000:
        bioactivity_class.append("inactive")
    elif float(i) <= 1000:
        bioactivity_class.append("active")
    else:
        bioactivity_class.append("intermediate")
```

Make molecule chembl id into lists and keep unique ones to reduce redundancy

```python
selection = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']    
df3 = df2[selection]
```

```python
df4 = pd.concat([df3, pd.Series(bioactivity_class, name='bioactivity_class')], axis=1)
df4
```

```python
df4.to_csv('coronavirus_preprocessed.csv', index=False)
```
